```python

u_egr_interp = interp1d(np.linspace(0, 60, 6000), u_egr_data[:6000], fill_value="extrapolate")
u_vgt_interp = interp1d(np.linspace(0, 60, 6000), u_vgt_data[:6000], fill_value="extrapolate")

global xr_list
global T1_list


global xr_list_save
global T1_list_save


xr_list = [0.0505]
T1_list = [411.3786]

##Variables required for training pre-trained NNs
Wei_list_save = []
Wegr_list_save = []
Tem_list_save = []
Wc_list_save = []

##Variables for empirical formulae###
ηvol_list_save = []


xr_list_save = []
T1_list_save = []
Ψegr_list_save = []
Wt_list_save = []
Πc_list_save = []
Πt_list_save = []
Ptηm_list_save = []
Aegr_list_save = []
Pc_list_save = []
Πe_list_save = []
xp_list_save = []
xv_list_save = []
fegr_list_save = []
fuvgt_list_save = []
Φc_list_save = []


#We define a functions which gives the integer of every variable we want to measure for a specific value
def func3(t, y):


    u_egr_data_i = u_egr_interp(t)  # Get interpolated u_egr_data at time t
    u_vgt_data_i = u_vgt_interp(t)  # Get interpolated u_vgt_data at time t

    pim, pem, omeggat = y


    ηc = 0.5#np.random.uniform(0.2, 0.7364)

    def Πegr(pim, pem):

        if (pim/pem < Πegropt):
            return Πegropt
        elif (pim/pem > 1):
            return 1
        else:
            return pim/pem

    def fegr(u_egr_data_i):

        if (u_egr_data_i <= ( -cegr2/(2 * cegr1))):
            return (cegr1 * (u_egr_data_i**2) + cegr2 * u_egr_data_i + cegr3)
        else :
            return (cegr3 - (cegr2**2)/(4 * cegr1))




    def fvgt(u_vgt_data_i):
        return cf2 + cf1 * np.sqrt(np.maximum(0, 1 - ((u_vgt_data_i - cvgt2)/cvgt1)**2))



    #We define the parameters of the equations (1) - (8)
    Ψegr = 1 - ((1 - Πegr(pim, pem))/(1 - Πegropt) - 1)**2

    Aegr = Aegrmax_true * fegr(u_egr_data_i)

    Πc = pim/pamb

    ηvol = cvol1 * np.sqrt(pim) + cvol2 * np.sqrt(ne) + cvol3

    Ψc =  2 * cpa * Tamb * (Πc**(1 - 1/γa) - 1)/((Rc**2) * omeggat**2)

    Φc = np.sqrt(np.maximum(0, (1 - cψ1 * (Ψc - cψ2)**2)/cφ1)) + cφ2

    Wc = (pamb * pi * (Rc**3) * omeggat) * Φc/(Ra * Tamb)

    Wei = (ηvol * pim * ne * Vd)/(120 * Ra * Tim)

    Wf = (1e-6) * uδ * ne * ncyl/120

    Weo = Wf + Wei

    Πt = pamb/pem

    Πe = pem/pim


    qin = ((Wf * qHV) * (1 - xr_list[-1])/(Wei + Wf))
    xp = (1 + (qin * xcv)/(cva * T1_list[-1] * (rc**(γa - 1))))
    Te = (ηsc_true * (Πe**(1 - 1/γa)) * (rc**(1 - γa)) * (xp**(1/γa - 1)) * (qin * ((1 - xcv)/(cpa) + xcv/cva) + T1_list[-1] * (rc**(γa - 1))))
    xv = (1 + qin * (1 - xcv)/(cpa * ((qin * xcv/cva) + T1_list[-1] * (rc**(γa - 1)))))
    xr = ((Πe**(1/γa)) * (xp**(-1/γa))/(rc * xv) )
    T1 = (xr * Te + (1 - xr) * Tim)


    Tem = Tamb + (Te - Tamb) * np.exp(((-htot_true * pi * dpipe * lpipe * npipe)/(Weo * cpe)))

    Wegr = Aegr * pim * Ψegr/(np.sqrt(Tem * Re))

    Wt = Avgtmax_true * pem * np.sqrt(np.maximum(0, (1 - Πt**Kt))) * fvgt(u_vgt_data_i) / np.sqrt(Tem * Re)

    Pc = Wc * cpa * Tamb * (Πc**(1-1/γa) - 1)/ηc

    Ptηm = ηtm * Wt * cpe * Tem * (1 - Πt**(1 - 1/γe))

    i = len(xr_list)
    xr_list.append(xr)
    T1_list.append(T1)


    if xr_list_save == [] or (60/(6000-1))*len(xr_list_save)<=t:
      xr_list_save.append(xr)
      T1_list_save.append(T1)
      Wegr_list_save.append(Wegr)
      Ψegr_list_save.append(Ψegr)
      Wt_list_save.append(Wt)
      Πc_list_save.append(Πc)
      Πt_list_save.append(Πt)
      Tem_list_save.append(Tem)
      Ptηm_list_save.append(Ptηm)
      Aegr_list_save.append(Aegr)
      Pc_list_save.append(Pc)
      Πe_list_save.append(Πe)
      xp_list_save.append(xp)
      xv_list_save.append(xv)
      Wei_list_save.append(Wei)
      Wc_list_save.append(Wc)
      ηvol_list_save.append(ηvol)
      Φc_list_save.append(Φc)


    if t<=10:
      print(f"{t:.2f}| qin ={qin:.2f}| T1 = {T1:.2f}| xv = {xv:.2f}| xp = {xp:.2f}| xr = {xr:.2f}| Te= {Te:.2f}| Wei = {Wei:.2f}| Πe = {Πe:.2f}| pem = {pem:.2f}| pim = {pim:.2f}| Wegr = {Wegr:.2f}| Wt = {Wt:.2f}| Ψegr = {Ψegr:.2f} \n")


    #Equation 1
    dpim_t = (Ra * Tim) * (1/Vim) * (Wc + Wegr - Wei)

    #Equation 2
    dpem_t = (Re * Tem) * (1/Vem) * (Weo - Wt - Wegr)

    #Equation 3
    domegga_t = (Ptηm - Pc)/(Jt * omeggat)


    return [dpim_t, dpem_t, domegga_t]
```

```python
# Assuming pim_data_true_tensor is your input tensor
pim_min = pim_data_true_tensor.min()
pim_max = pim_data_true_tensor.max()

# Apply Min-Max Normalization to scale values between [0, 1]
pim_data_true_norm_tensor = (pim_data_true_tensor - pim_min) / (pim_max - pim_min)
```

```python
class FFNN1_PT(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN):
        super().__init__()
        activation_hidden = nn.Tanh
        #self.output_transformation = nn.Sigmoid()

        self.ffni = nn.Sequential(nn.Linear(N_INPUT, N_HIDDEN), activation_hidden())
        self.ffnh = nn.Sequential(nn.Linear(N_HIDDEN, N_HIDDEN), activation_hidden())
        self.ffne = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.ffni(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffne(x)
        #x = self.output_transformation(x)
        return x
```

```python
torch.manual_seed(123)
PINN1_PT = FFNN1_PT(1, 1, 64)
```

```python
def NN_PT_loss(model, x_data, y_data):

    y_pred = model(x_data)

    return torch.mean((y_data - y_pred)**2)
```

```python
optimiser_N1_PT = optim.Adam(PINN1_PT.parameters(), lr=0.0001)
```

```python
# ---------------------------------------
# 6. Training Loop
# ---------------------------------------
num_epochs_N1_PT = 10000
print_every_N1_PT = 1000
loss_NN1_PT_list = []

PINN1_PT.train()
for epoch in range(num_epochs_N1_PT):
    optimiser_N1_PT.zero_grad()
    loss_NN1_PT = NN_PT_loss(PINN1_PT, pim_data_true_norm_tensor, eta_vol_data_tensor)
    loss_NN1_PT_list.append(loss_NN1_PT.item())
    loss_NN1_PT.backward()
    optimiser_N1_PT.step()

    if (epoch+1) % print_every_N1_PT == 0:
        print(f"Epoch {epoch+1}/{num_epochs_N1_PT}, "
              f"Total Loss = {loss_NN1_PT.item():.7f}")

plt.plot(np.linspace(0, num_epochs_N1_PT, len(loss_NN1_PT_list)), loss_NN1_PT_list)
plt.xlabel('Epoch')
plt.ylabel('Loss_N1_PT')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# ---------------------------------
# 1. Prepare Data
# ---------------------------------

# Convert to tensors
Wei_data_true_tensor = torch.tensor(Wei_data_true, dtype=torch.float32).view(-1, 1)
pim_data_true_tensor = torch.tensor(pim_data_true, dtype=torch.float32).view(-1, 1)

eta_vol_data = 120 * Wei_data_true * Ra * Tim / (pim_data_true * ne * Vd)
eta_vol_data_tensor = torch.tensor(eta_vol_data, dtype=torch.float32).view(-1, 1)

# Normalize input data
pim_min = pim_data_true_tensor.min()
pim_max = pim_data_true_tensor.max()
pim_data_true_norm_tensor = (pim_data_true_tensor - pim_min) / (pim_max - pim_min)

# ---------------------------------
# 2. Train-Test Split (80%-20%)
# ---------------------------------
(
    pim_train, pim_test,
    eta_train, eta_test
) = train_test_split(pim_data_true_norm_tensor, eta_vol_data_tensor, test_size=0.2, random_state=42)

# Convert to PyTorch tensors
pim_train_tensor = torch.tensor(pim_train, dtype=torch.float32)
eta_train_tensor = torch.tensor(eta_train, dtype=torch.float32)
pim_test_tensor = torch.tensor(pim_test, dtype=torch.float32)
eta_test_tensor = torch.tensor(eta_test, dtype=torch.float32)

# ---------------------------------
# 3. Define Neural Network
# ---------------------------------
class FFNN1_PT(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN):
        super().__init__()
        activation_hidden = nn.Tanh
        self.ffni = nn.Sequential(nn.Linear(N_INPUT, N_HIDDEN), activation_hidden())
        self.ffnh = nn.Sequential(nn.Linear(N_HIDDEN, N_HIDDEN), activation_hidden())
        self.ffne = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.ffni(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffne(x)
        return x

torch.manual_seed(123)
PINN1_PT = FFNN1_PT(1, 1, 64)

# ---------------------------------
# 4. Loss Function & Optimizer
# ---------------------------------
def NN1_PT_loss(model, x_data, y_data):
    y_pred = model(x_data)
    return torch.mean((y_data - y_pred)**2)

optimiser_N1_PT = optim.Adam(PINN1_PT.parameters(), lr=0.0001)

# ---------------------------------
# 5. Training Loop
# ---------------------------------
num_epochs_N1_PT = 10000
print_every_N1_PT = 1000
train_losses_N1_PT = []
test_losses_N1_PT = []

PINN1_PT.train()
for epoch in range(num_epochs_N1_PT):
    optimiser_N1_PT.zero_grad()

    # Compute training loss
    train_loss_N1_PT = NN1_PT_loss(PINN1_PT, pim_train_tensor, eta_train_tensor)
    train_losses_N1_PT.append(train_loss_N1_PT.item())

    train_loss_N1_PT.backward()
    optimiser_N1_PT.step()

    # Compute test loss (without backpropagation)
    PINN1_PT.eval()
    with torch.no_grad():
        test_loss = NN_PT_loss(PINN1_PT, pim_test_tensor, eta_test_tensor)
        test_losses_N1_PT.append(test_loss.item())

    PINN1_PT.train()  # Switch back to training mode

    if (epoch+1) % print_every_N1_PT == 0:
        print(f"Epoch {epoch+1}/{num_epochs_N1_PT}, "
              f"Train Loss = {train_loss_N1_PT.item():.7f}, "
              f"Test Loss = {test_loss.item():.7f}")

# ---------------------------------
# 6. Plot Train vs Test Loss
# ---------------------------------
plt.figure(figsize=(8, 5))
plt.plot(np.linspace(0, num_epochs_N1_PT, len(train_losses_N1_PT)), train_losses_N1_PT, label="Train Loss")
plt.plot(np.linspace(0, num_epochs_N1_PT, len(test_losses_N1_PT)), test_losses_N1_PT, label="Test Loss", linestyle="dashed")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------------
# 7. Evaluate the Trained Model
# ---------------------------------
PINN1_PT.eval()
t_plot = np.linspace(0, 60, 300).reshape(-1, 1).astype(np.float32)
t_plot_tensor = torch.tensor(t_plot, dtype=torch.float32)

eta_vol_pred_plot = PINN1_PT(pim_data_true_norm_tensor).detach().numpy()

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(t_plot, eta_vol_data, 'k--', label='Lab Data')
plt.plot(t_plot, eta_vol_pred_plot, 'b', label='PINN Prediction')
plt.xlabel('t')
plt.ylabel('h_vol')
plt.legend()
plt.grid(True)
plt.show()
```

#$N_1(x; \theta_1) - predicting\ pim,pem$

```python
class FFN1(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN):
        super().__init__()
        activation_hidden = nn.Tanh
        self.output_transformation = nn.Softplus()  # Use Softplus instead of Sigmoid

        # Define the layers
        self.ffni = nn.Sequential(nn.Linear(N_INPUT, N_HIDDEN), activation_hidden())
        self.ffnh = nn.Sequential(nn.Linear(N_HIDDEN, N_HIDDEN), activation_hidden())
        self.ffne = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.ffni(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffne(x)

        # Apply Softplus transformation correctly
        p1 = self.output_transformation(x[:, 0]) + 0.5
        p2 = self.output_transformation(x[:, 1])

        # Scale the outputs
        output1 = p1 * 1e5
        output2 = p2 * 1e5

        return torch.stack([output1, output2], dim=1)
```

```python
class FFN2(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN):
        super().__init__()
        activation_hidden = nn.Tanh
        self.output_transformation = nn.Softplus()

        # Define the layers
        self.ffni = nn.Sequential(nn.Linear(N_INPUT, N_HIDDEN), activation_hidden())
        self.ffnh = nn.Sequential(nn.Linear(N_HIDDEN, N_HIDDEN), activation_hidden())
        self.ffne = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.ffni(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffne(x)

        # Apply Sigmoid activation on the output
        x = self.output_transformation(x)

        # Apply scaling (multiply by 100)
        return  0.03 * x
```

```python
import torch
import torch.optim as optim
import numpy as np

pinn1 = FFN1(1, 2, 64)
pinn2 = FFN2(1, 1, 64)
pinn3 = FFN3(1, 1, 64)
pinn5 = FFN5(1, 1, 64)

# Define learnable parameters
Aegrmax = torch.nn.Parameter(torch.tensor([1], dtype=torch.float32, requires_grad=True))
hsc = torch.nn.Parameter(torch.tensor([1], dtype=torch.float32, requires_grad=True))
htot = torch.nn.Parameter(torch.tensor([1], dtype=torch.float32, requires_grad=True))
Avgtmax = torch.nn.Parameter(torch.tensor([1], dtype=torch.float32, requires_grad=True))

# Store parameter values over epochs
Aegrmax_values, hsc_values, htot_values, Avgtmax_values = [], [], [], []

# Define optimizer
optimizer = optim.Adam(
    list(pinn1.parameters()) + list(pinn2.parameters()) + list(pinn3.parameters()) +
     list(pinn5.parameters()) +
    [Aegrmax, hsc, htot, Avgtmax],
    lr=0.001
)
```

```python
Loss_history_N1 = []
Loss_history_N2 = []
Loss_history_N3 = []
Loss_history_N4 = []
Loss_history_N5 = []
Loss_history_N6 = []
Loss_history_Wegr = []
Loss_history_total = []

num_epochs = 21000
print_every = 1000

for epoch in range(num_epochs):
    optimizer.zero_grad()

    ####################################################
    # -----------Initial Condition Losses -------------#
    ####################################################


    #IC Losses for pim,pem
    p_pred_boundary = pinn1(t_boundary)
    pim_pred_boundary = p_pred_boundary[:,0].view(-1,1)
    pem_pred_boundary = p_pred_boundary[:,1].view(-1,1)

    IC_loss_pim =  torch.mean((torch.squeeze(pim_pred_boundary) - 8.0239e4)**2)
    IC_loss_pem =  torch.mean((torch.squeeze(pem_pred_boundary) - 8.1220e4)**2)

    #IC Losses for omeggat
    omeggat_pred_boundary = pinn5(t_boundary).view(-1,1)

    IC_loss_omeggat =  torch.mean((torch.squeeze(omeggat_pred_boundary) - 1.5827e3)**2)

    #IC Losses for T1
    T1_pred_boundary = pinn3(t_boundary).view(-1,1)

    IC_loss_T1 = torch.mean((torch.squeeze(T1_pred_boundary) - 318.5961)**2)

    #IC Losses for xr
    xr_pred_boundary = pinn2(t_boundary).view(-1,1)

    IC_loss_xr =  torch.mean((torch.squeeze(xr_pred_boundary) - 0.02995)**2)

    IC_Loss_Total = IC_loss_pim + IC_loss_pem + IC_loss_omeggat + IC_loss_T1 + IC_loss_xr



    ####################################################
    # -----------       Data Losses       -------------#
    ####################################################

    #Data Loss for pim,pem
    p_pred_data = pinn1(t_data)
    pim_pred_data = p_pred_data[:,0].view(-1,1)
    pem_pred_data = p_pred_data[:,1].view(-1,1)

    Data_loss_pim =  torch.mean((torch.from_numpy(pim_data_true) - torch.squeeze(pim_pred_data))**2)
    Data_loss_pem = torch.mean((torch.from_numpy(pem_data_true) - torch.squeeze(pem_pred_data))**2)

    #Data Loss for omeggat
    omeggat_pred_data = pinn5(t_data)
    omeggat_pred_data = omeggat_pred_data.view(-1,1)

    Data_loss_omeggat = torch.mean((torch.from_numpy(omeggat_data_true) - torch.squeeze(omeggat_pred_data))**2)



    #Data Loss for Wegr

    T1_pred_data = pinn3(t_data)
    T1_pred_data = T1_pred_data.view(-1,1)

    xr_pred_data = pinn2(t_data)
    xr_pred_data = xr_pred_data.view(-1,1)

    Pegr = Πegr(pim_pred_data, pem_pred_data, Πegropt).view(-1, 1)
    h_vol_data = PINN1_PT(pim_pred_data).view(-1, 1)

    Aegr = (Aegrmax) * fegr_data
    C_egr_data = 1 - ((1 - Pegr) / (1 - Πegropt) - 1)**2

    pem_pred_data_min = pem_pred_data.min()
    pem_pred_data_max = pem_pred_data.max()
    pem_pred_data_norm = (pem_pred_data - pem_pred_data_min) / (pem_pred_data_max - pem_pred_data_min)


    # Avoid division by zero
    Wei_data = h_vol_data * pim_pred_data * ne * Vd / (120 * Ra * (Tim ))
    Wf = (1e-6) * uδ * ne * ncyl/120
    Weo_data = Wf + Wei_data
    Pe_data = pem_pred_data / (pim_pred_data)


    qin_data = ((Wf * qHV) * (1 - xr_pred_data) / (Wei_data + Wf ))
    xp_data = (1 + (qin_data * xcv) / (cva * T1_pred_data * (rc**(γa - 1))))
    xv_data = (1 + qin_data * (1 - xcv) / (cpa * ((qin_data * xcv / cva) + T1_pred_data * (rc**(γa - 1)))))

    Te_data = ((hsc) * (Pe_data**(1 - 1/γa)) * (rc**(1 - γa)) *
               (xp_data**(1/γa - 1)) * (qin_data * ((1 - xcv) / cpa + xcv / cva) + T1_pred_data * (rc**(γa - 1))))


    exp_term = (-(htot)* pi * dpipe * lpipe * npipe) / (Weo_data * cpe )

    Tem_data = Tamb + (Te_data - Tamb) * torch.exp(exp_term)


    Wegr_data = (Aegr * pim_pred_data * C_egr_data) / (torch.sqrt(Tem_data * Re ))



    # Compute Wegr loss
    Data_loss_Wegr = torch.mean((torch.squeeze(Wegr_data) - torch.from_numpy(Wegr_data_true))**2)


    # Total Data Loss
    Data_Loss_Total =  Data_loss_pim +  Data_loss_pem +  Data_loss_omeggat +  Data_loss_Wegr
    torch.autograd.set_detect_anomaly(True)

    ####################################################
    # -----------      Physics Losses     -------------#
    ####################################################

    #pim,pem
    p_pred_physics = pinn1(t_physics)
    pim_pred_physics = p_pred_physics[:,0].view(-1,1)
    pem_pred_physics = p_pred_physics[:,1].view(-1,1)

    du_pim_pred_physics_dt = torch.autograd.grad(pim_pred_physics, t_physics, torch.ones_like(pim_pred_physics), create_graph = True, allow_unused= True)[0]
    du_pem_pred_physics_dt = torch.autograd.grad(pem_pred_physics, t_physics, torch.ones_like(pem_pred_physics), create_graph = True, allow_unused= True)[0]

    #omeggat
    omeggat_pred_physics = pinn5(t_physics)
    omeggat_pred_physics = omeggat_pred_physics.view(-1,1)

    du_omeggat_pred_physics_dt = torch.autograd.grad(omeggat_pred_physics, t_physics, torch.ones_like(omeggat_pred_physics), create_graph = True, allow_unused= True)[0]

    #T1
    T1_pred_physics = pinn3(t_physics)
    T1_pred_physics = T1_pred_physics.view(-1,1)


    #xr
    xr_pred_physics = pinn2(t_physics)
    xr_pred_physics = xr_pred_physics.view(-1,1)


    #-----Functions quivaltent to the pre-trained NNs-----#

    pim_pred_physics_min = pim_pred_physics.min()
    pim_pred_physics_max = pim_pred_physics.max()
    pim_pred_physics_norm = (pim_pred_physics - pim_pred_physics_min) / (pim_pred_physics_max - pim_pred_physics_min)

    pem_pred_physics_min = pem_pred_physics.min()
    pem_pred_physics_max = pem_pred_physics.max()
    pem_pred_physics_norm = (pem_pred_physics - pem_pred_physics_min) / (pem_pred_physics_max - pem_pred_physics_min)

    omeggat_pred_physics_min = omeggat_pred_physics.min()
    omeggat_pred_physics_max = omeggat_pred_physics.max()
    omeggat_pred_physics_norm = (omeggat_pred_physics - omeggat_pred_physics_min) / (omeggat_pred_physics_max - omeggat_pred_physics_min)

    Pt = pamb/pem_pred_physics

    Pc = pim_pred_physics/pamb

    #h_vol
    h_vol_physics = PINN1_PT(pim_pred_physics_norm)
    h_vol_physics = h_vol_physics.view(-1,1)

    #Fvgt,Πt
    fvgt_Pit_physics = PINN3_PT(torch.cat((pem_pred_physics_norm, Pt), dim=1))
    fvgt_Pit_physics = fvgt_Pit_physics.view(-1,1)

    #Φc
    Φc_physics = PINN4_PT(torch.cat((Tamb_tensor_physics, Pc, omeggat_pred_physics_norm), dim=1))
    Φc_physics = Φc_physics.view(-1,1)

    #----Calculating all the parameters for the losses----#
    Wei = h_vol_physics * pim_pred_physics * ne * Vd/(120 * Ra * Tim)
    Wf = (1e-6) * uδ * ne * ncyl/120
    Weo = Wf + Wei
    Wc = (pamb * 3.14 * (Rc**3) * omeggat_pred_physics * Φc_physics)/(Ra * Tamb)

    Pe = pem_pred_physics/pim_pred_physics
    qin = ((Wf * qHV) * (1 - xr_pred_physics)/(Wei + Wf))
    xp = (1 + (qin * xcv)/(cva * T1_pred_physics * (rc**(γa - 1))))
    xv = (1 + qin * (1 - xcv)/(cpa * ((qin * xcv/cva) + T1_pred_physics * (rc**(γa - 1)))))
    Te = ((hsc) * (Pe**(1 - 1/γa)) * (rc**(1 - γa)) * (xp**((1/γa) - 1)) * (qin * ((1 - xcv)/(cpa) + xcv/cva) + T1_pred_physics * (rc**(γa - 1))))
    Tem = Tamb + (Te - Tamb) * torch.exp(((-(htot) * pi * dpipe * lpipe * npipe)/(Weo * cpe)))

    Pegr = Πegr(pim_pred_physics, pem_pred_physics, Πegropt)
    Pegr = Pegr.view(-1,1)
    Aegr = (Aegrmax_true) * fegr_physics
    C_egr = 1 - ((1 - Pegr)/(1 - Πegropt) - 1)**2
    Wegr = (Aegr * pim_pred_physics * C_egr)/(torch.sqrt(Tem * Re))

    Wt = Avgtmax_true * pem_pred_physics * fvgt_Pit_physics/(torch.sqrt(Tem * Re))
    Pthm = ηtm * Wt * cpe * Tem * (1 - Pt**(1 - 1/γa))


    #--------------------------------------Computing the equation losses-----------------------------------------#

    Eq_loss_pim = (1e-10) *torch.mean((du_pim_pred_physics_dt - (Ra * Tim/Vim) * (Wc + Wegr - Wei) )**2)
    Eq_loss_pem = (1e-10) *torch.mean((du_pem_pred_physics_dt - (Re * Tem/Vem) * (Weo - Wt - Wegr) )**2)
    Eq_loss_omeggat = (1e-6) *torch.mean((du_omeggat_pred_physics_dt - (Pthm - Pc)/(Jt * omeggat_pred_physics ))**2)
    Eq_loss_T1 =  (1e2) * torch.mean((T1_pred_physics - ((xr_pred_physics * Te) + (1 - xr_pred_physics) * (Tim)))**2)
    Eq_loss_xr =  (1e4) * torch.mean((xr_pred_physics - ((Pe**(1/γa)) * (xp**(-1/γa))/(rc * xv)))**2)

    Eq_loss_Total = Eq_loss_T1 + Eq_loss_xr

    Total_Loss = IC_Loss_Total + Data_Loss_Total
    Loss_history_total.append(Total_Loss.item())

    Total_Loss.backward()
    optimizer.step()

    Avgtmax_values.append(Avgtmax.item())
    Aegrmax_values.append(Aegrmax.item())
    hsc_values.append(hsc.item())
    htot_values.append(htot.item())


    if (epoch+1) % print_every == 0:
        print(f"Epoch {epoch+1}/{num_epochs},"
              f"Eq_loss_pim = {Eq_loss_pim}| Eq_loss_pem = {Eq_loss_pem}| Eq_loss_omeggat = {Eq_loss_omeggat}|  Eq_loss_T1 = {Eq_loss_T1} | Eq_loss_xr = {Eq_loss_xr} ,\n")
        print(f"Epoch {epoch+1}/{num_epochs},"
              f"Data_loss_pim = {Data_loss_pim}| Data_loss_pem = {Data_loss_pem}| Data_loss_omeggat = {Data_loss_omeggat}| Data_loss_Wegr = {Data_loss_Wegr}")
        print(f"Epoch {epoch+1}/{num_epochs},"
              f"IC_loss_pim = {IC_loss_pim}| IC_loss_pem = {IC_loss_pem}| IC_loss_omeggat = {IC_loss_omeggat}| IC_loss_T1 = {IC_loss_T1} | IC_loss_xr = {IC_loss_xr} ,\n")
        print(f"Total_loss = {Total_Loss}")
        print("\n")
    if (epoch) % 5000 == 0:
        plt.plot(np.linspace(0,60,300), omeggat_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), omeggat_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("omeggat")
        plt.show()

        plt.plot(np.linspace(0,60,300), T1_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), T1_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("T1")
        plt.show()

        plt.plot(np.linspace(0,60,300), u_egr1_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), u_egr1_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("u_egr1")
        plt.show()

        plt.plot(np.linspace(0,60,300), u_egr2_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), u_egr2_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("u_egr2")
        plt.show()

        plt.plot(np.linspace(0,60,300), pim_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), pim_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt
        plt.xlabel("time")
        plt.ylabel("pim")
        plt.show()

        plt.plot(np.linspace(0,60,300), pem_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), pem_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()

        plt.xlabel("time")
        plt.ylabel("pem")
        plt.show()

        plt.plot(np.linspace(0,60,300), u_vgt_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), u_vgt_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("u_vgt")
        plt.show()

        plt.plot(np.linspace(0,60,300), xr_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), xr_pred_physics.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("xr")
        plt.show()

```

```python

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Swish activation
def swish(x):
    return x * torch.sigmoid(x)

# Define MLP for Branch and Trunk networks
class MLP(nn.Module):
    def __init__(self, input_dim, output_dim, width=128, depth=3):
        super(MLP, self).__init__()
        layers = [nn.Linear(input_dim, width), nn.SiLU()]
        for _ in range(depth - 2):
            layers.append(nn.Linear(width, width))
            layers.append(nn.SiLU())
        layers.append(nn.Linear(width, output_dim))
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)

# Define DeepONet
class DeepONet(nn.Module):
    def __init__(self):
        super(DeepONet, self).__init__()

        # Branch Networks
        self.branch1 = MLP(1, 128)
        self.branch2 = MLP(1, 128)
        self.branch3 = MLP(5, 128)  # Adjusted for initial conditions
        self.branch4 = MLP(1, 128)
        self.branch5 = MLP(1, 128)
        self.branch6 = MLP(1, 128)
        self.branch7 = MLP(1, 128)

        # Trunk Networks
        self.trunk1 = MLP(1, 128)
        self.trunk2 = MLP(1, 128)
        self.trunk3 = MLP(1, 128)
        self.trunk4 = MLP(1, 128)
        self.trunk5 = MLP(1, 128)

    def forward(self, t, uegr, uvgt, initial_conditions):
        initial_conditions = initial_conditions.expand(300, -1)  # Ensure correct batch expansion

        # Compute embeddings
        b1 = self.branch1(t)
        b2 = self.branch2(t)
        b3 = self.branch3(initial_conditions)
        b4 = self.branch4(uegr)
        b5 = self.branch5(uvgt)
        b6 = self.branch6(uegr)
        b7 = self.branch7(uvgt)

        t1 = self.trunk1(t)
        t2 = self.trunk2(t)
        t3 = self.trunk3(t)
        t4 = self.trunk4(t)
        t5 = self.trunk5(t)

        # Compute output states
        Pim = torch.sum(b1 * t1, dim=1, keepdim=True)
        Pem = torch.sum(b2 * t2, dim=1, keepdim=True)
        xr = torch.sum(b3 * t3, dim=1, keepdim=True)
        T1 = torch.sum(b4 * t4, dim=1, keepdim=True)
        omega_t = torch.sum(b5 * t5, dim=1, keepdim=True)

        uegr_pred = torch.sum(b6 * t4, dim=1, keepdim=True)
        uvgt_pred = torch.sum(b7 * t5, dim=1, keepdim=True)

        return Pim, Pem, xr, T1, omega_t, uegr_pred, uvgt_pred

# Initialize model
deepONet = DeepONet()

# Define learnable parameters
Aegrmax = nn.Parameter(torch.tensor([1e-4], dtype=torch.float32, requires_grad=True))
hsc = nn.Parameter(torch.tensor([1], dtype=torch.float32, requires_grad=True))
htot = nn.Parameter(torch.tensor([1], dtype=torch.float32, requires_grad=True))
Avgtmax = nn.Parameter(torch.tensor([1e-4], dtype=torch.float32, requires_grad=True))
initial_conditions = torch.tensor([
    [8.0239e4, 8.1220e4, 1.5827e3, 318.5961, 0.02995]  # Correct values for Pim, Pem, ωt, T1, xr
], dtype=torch.float32).expand(300, 5)  # Expanding to match batch size

# Define optimizer
optimizer = optim.Adam(
    list(deepONet.parameters()) + [Aegrmax, hsc, htot, Avgtmax], lr=0.001
)

# Training loop
num_epochs = 100000
print_every = 1000


for epoch in range(num_epochs):
    optimizer.zero_grad()

    # Forward pass
    Pim, Pem, xr, T1, omega_t, uegr_pred, uvgt_pred = deepONet(t_data, uegr_data_true_tensor, uvgt_data_true_tensor, initial_conditions)

    # Compute losses
    IC_loss_pim = torch.mean((Pim - 8.0239e4) ** 2)
    IC_loss_pem = torch.mean((Pem - 8.1220e4) ** 2)
    IC_loss_omeggat = torch.mean((omega_t - 1.5827e3) ** 2)
    IC_loss_T1 = torch.mean((T1 - 318.5961) ** 2)
    IC_loss_xr = torch.mean((xr - 0.02995) ** 2)
    IC_Loss_Total = IC_loss_pim + IC_loss_pem + IC_loss_omeggat + IC_loss_T1 + IC_loss_xr

    # Data loss
    Data_loss_pim = torch.mean((Pim - pim_data_true_tensor) ** 2)
    Data_loss_pem = torch.mean((Pem - pem_data_true_tensor) ** 2)
    Data_loss_omeggat = torch.mean((omega_t - omeggat_data_true_tensor) ** 2)
    Data_loss_Wegr = torch.mean((uegr_pred - Wegr_data_true_tensor) ** 2)
    Data_Loss_Total = Data_loss_pim + Data_loss_pem + Data_loss_omeggat + Data_loss_Wegr

    # Total loss
    Total_Loss = IC_Loss_Total + Data_Loss_Total
    Total_Loss.backward()
    optimizer.step()

    if (epoch + 1) % print_every == 0:
        print(f"Epoch {epoch+1}/{num_epochs}, Total Loss: {Total_Loss.item()}")
    if (epoch) % 5000 == 0:
        plt.plot(np.linspace(0,60,300), omeggat_data_true, color = "black", label = "True")
        plt.plot(np.linspace(0,60,300), omega_t.detach().numpy(), color = "red", linestyle = "dashed", label = "Predicted")
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("omeggat")
        plt.show()

```

```python
import torch
import torch.nn as nn

class FFN1(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN):
        super().__init__()
        activation_hidden = nn.Tanh
        self.output_transformation = nn.Softplus()  # Use Softplus instead of Sigmoid

        # Define the layers
        self.ffni = nn.Sequential(nn.Linear(N_INPUT, N_HIDDEN), activation_hidden())
        self.ffnh = nn.Sequential(nn.Linear(N_HIDDEN, N_HIDDEN), activation_hidden())
        self.ffne = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.ffni(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffnh(x)
        x = self.ffne(x)

        # Apply Softplus transformation correctly
        p1 = self.output_transformation(x[:, 0]) + 0.5
        p2 = self.output_transformation(x[:, 1])

        # Scale the outputs
        output1 = p1 * 1e5
        output2 = p2 * 1e5

        return torch.stack([output1, output2], dim=1)
```
